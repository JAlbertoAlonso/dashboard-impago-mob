```python
from utils_impago_ondemand import ImpagoOnDemandEngine, Scenario

import pandas as pd
```

```python
# Configuración
df_master_dummy = pd.read_csv('data/df_master_dummy.csv')
```

# Escenario Base A

- Mora +4
- Sin filtros
- Sin desagregados
- Con castigos

```python
S0A = Scenario(
    name="S0A_base_BGI4",
    tipo_mora="BGI4+",
    metric_mode="exposure",
    filters={}
)
engine_off = ImpagoOnDemandEngine(
    df_master_dummy,
    castigo_enabled=False,               # OFF
    debug=False
)
res_S0A = engine_off.run_scenario(S0A, show=False, save_outputs=True)
{k: res_S0A[k] for k in [
    "castigo_enabled", "rows_before_castigo", "rows_after_castigo",
    "folios_before_castigo", "folios_after_castigo"
]}
res_S0A["matrix_display"]
```
