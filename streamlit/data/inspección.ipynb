```python
df_1 = pd.read_csv('cosechas_dimensiones_20260301.csv')

mes_map = {
    'ene': '01', 'feb': '02', 'mar': '03', 'abr': '04',
    'may': '05', 'jun': '06', 'jul': '07', 'ago': '08',
    'sep': '09', 'oct': '10', 'nov': '11', 'dic': '12'
}

def convertir_cosecha(c):
    try:
        mes, anio = c.split('-')
        mes_num = mes_map[mes.lower()]
        anio_full = f"20{anio}"  
        return f"{anio_full}-{mes_num}"
    except:
        return None  

df_1['cosecha'] = df_1['cosecha'].astype(str).apply(convertir_cosecha)
```

```python
diagnostico = df_1.groupby("folio")["cosecha"].nunique()

print(diagnostico.describe())
print("Folios con más de una cosecha:", (diagnostico > 1).sum())
```

```python
# Guardar la cosecha observada original
df_1["cosecha_obs"] = df_1["cosecha"]

# Normalizar tipos
df_1["cosecha_obs"] = pd.to_datetime(df_1["cosecha_obs"], format="%Y-%m", errors="coerce")
df_1["MOB"] = pd.to_numeric(df_1["MOB"], errors="coerce")

# Reconstruir cosecha fija de originación
def reconstruir_cosecha_origen(row):
    fecha = row["cosecha_obs"]
    mob = row["MOB"]

    if pd.isna(fecha) or pd.isna(mob):
        return pd.NA

    mob = int(mob)
    return (fecha.to_period("M") - (mob - 1)).strftime("%Y-%m")

df_1["cosecha_origen"] = df_1.apply(reconstruir_cosecha_origen, axis=1)

diagnostico_origen = df_1.groupby("folio")["cosecha_origen"].nunique()

print(diagnostico_origen.describe())
print("Folios con más de una cosecha_origen:", (diagnostico_origen > 1).sum())
```

```python
df_1 = df_1.rename(columns={"cosecha": "cosecha_obs"})
df_1 = df_1.rename(columns={"cosecha_origen": "cosecha"})
```
